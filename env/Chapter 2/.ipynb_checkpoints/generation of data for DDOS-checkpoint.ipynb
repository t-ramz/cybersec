```python
# IP address generation found at https://stackoverflow.com/questions/21014618/python-randomly-generated-ip-address-as-string
import time
import random
import socket
import struct
protocol = ["IGMP", "UDP", "TCP", "DNS"]
myData = []
for i in range(10, 1000):
    myData.append("\n" + str(i))
    myTime = int(time.time() - random.randint(29000000, 79000000)) # generate a time from about a year ago to the time the book was written
    myData.append(", " + str(myTime))
    mySIP = socket.inet_ntoa(struct.pack('>I', random.randint(1, 0xffffffff))) # generate random IP addresses
    myData.append(", " + str(mySIP))
    myDIP = socket.inet_ntoa(struct.pack('>I', random.randint(1, 0xffffffff)))
    myData.append(", " + str(myDIP))
    myVolume = random.randint(1, 50)
    myData.append(", " + str(myVolume))
    myProtocol = protocol[random.randint(0, 3)]
    myData.append(", " + myProtocol)
myData
```
