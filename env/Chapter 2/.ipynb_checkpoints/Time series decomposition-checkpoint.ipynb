```python
import pandas
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
air = pandas.read_csv('airline_data.csv', index_col=0, parse_dates=True, header=0)
result = seasonal_decompose(air, model='additive')
result.plot()
pyplot.show()
```
